```python
import string

from collections import Counter

import matplotlib.pyplot as plt
from wordcloud import WordCloud
from PIL import Image
import numpy as np
from collections import Counter

# Load text file
!pip install requests

import requests

target_url ='http://www.gutenberg.org/files/2265/2265.txt'

response = requests.get(target_url)

data = response.text

# Remove punctuation and convert to lowercase
text = data.translate(str.maketrans('', '', string.punctuation)).lower()

# Split text into words
words = text.split()

# Count word frequencies
word_counts = Counter(words)

# Print top 10 most common words
print(word_counts.most_common(10))

# Create bar chart
top_words = dict(word_counts.most_common(20))
plt.bar(top_words.keys(), top_words.values())
plt.xticks(rotation=90)
plt.title('Top 20 Word Frequencies in Hamlet')
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.show()


# Load mask image
import imageio
image_file="https://media.cheggcdn.com/media/216/21621ee5-e80f-47f3-9145-513f2229b390/phploeBuh.png"
mask_image = imageio.imread(image_file)

# Create word cloud
mask = np.array(mask_image)
word_cloud = WordCloud(background_color='white', mask=mask).generate(text)

# Plot word cloud
plt.figure(figsize=(10, 10))
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()



```
